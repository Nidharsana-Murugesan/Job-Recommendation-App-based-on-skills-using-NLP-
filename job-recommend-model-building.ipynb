```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
from fuzzywuzzy import process

def get_encoded_value(input_str, column_name):
    # Ensure column exists in the encoding_dict
    if column_name not in encoding_dict:
        raise ValueError(f"Column '{column_name}' not found in encoding_dict.")
    
    # Retrieve the dictionary that maps encoded values to original strings
    column_dict = encoding_dict[column_name]  # This should map encoded integers -> original strings
    
    # Reverse the dictionary to map strings -> encoded values for fuzzy matching
    reversed_dict = {str(v): k for k, v in column_dict.items()}  # String -> Encoded Integer
    
    # Find the closest string match using fuzzy matching on the original string values
    closest_match = process.extractOne(input_str, reversed_dict.keys())  # Fuzzy match on strings
    
    # Return the encoded value corresponding to the closest matched string
    return reversed_dict[closest_match[0]]  # Return the encoded integer

# Example usage:
user_input = "mid level"
encoded_job_title = get_encoded_value(user_input, 'Job Title')
print(f"Encoded value for '{user_input}' is: {encoded_job_title}")
```

```python
def preprocess_text(text):
    # Ensure that the input is a string
    if not isinstance(text, str):
        return ""
    
    # Lowercasing
    text = text.lower()
    
    # Removing non-alphabetical characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Tokenization
    tokens = text.split()
    
    # Removing stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Return the processed text
    processed_text = ' '.join(tokens)

    # Debugging: Print processed text for each job description
    # print(f"Processed text: '{processed_text}'")  # Debugging line

    return processed_text
```
